# Capstone Project: Modelling Train and Test

```python
# import necessary libraries

import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import PolynomialFeatures, PowerTransformer, OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, make_scorer, recall_score, precision_recall_curve, auc, precision_score, recall_score, f1_score, roc_auc_score, accuracy_score
from sklearn.svm import LinearSVC, SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, VotingClassifier

from scipy import stats
import statsmodels.stats.proportion as smprop
```

```python
# read in data

fashion = pd.read_csv('data/for_modelling.csv')
fashion.head()
```

### Model Preparation

#### Features matrix and target variable

```python
# features matrix and target variable

features = fashion.drop(columns = 'EI')
X = features
y = fashion['EI']
```

#### Separate categorical and numerical columns

```python
cat_cols = X.select_dtypes(include='object').columns
num_cols = X.select_dtypes(include=np.number).columns
```

#### Train-Test-Split

```python
# train-test-split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Modelling and Tuning

```python
# set up transformer for encoding and scaling

preprocessor = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(), cat_cols),
    ('num', StandardScaler(), num_cols)])
```

```python
# get scores function

def get_scores(gs, X, y):
    pred = gs.predict(X)
    print(f'Accuracy score: {round(accuracy_score(y, pred), 4)}')
    print(f'Precision score: {round(precision_score(y, pred, average="macro"), 4)}')
    print(f'Recall score: {round(recall_score(y, pred, average="macro"), 4)}')
    print(f'F1 score: {round(f1_score(y, pred, average="macro"), 4)}')
```

```python
# function for confusion matrix

def get_cm(gs, X, y):
    pred = gs.predict(X)
    cm = confusion_matrix(y, pred)
    cm_plot = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['XS', 'S', 'MS', 'NS', 'XNS'])
    cm_plot.plot()
```

#### k-Nearest Neighbors

```python
# instatiate pipeline

knn_pipe = Pipeline([('preprocessor', preprocessor),
                            ('knn', KNeighborsClassifier())])
```

```python
# setting parameters

knn_params = {'knn__n_neighbors': [5, 6, 7, 8, 9, 10, 11],
              'knn__metric': ['manhattan', 'minkowski'],
              'knn__weights': ['distance', 'uniform']}
```

```python
# instantiate gridsearch

gs1 = GridSearchCV(knn_pipe, param_grid = knn_params, cv = 5)
```

```python
# train model

gs1.fit(X_train, y_train)
```

```python
# get best hyperparams

knn_best_params = gs1.best_params_
knn_best_model = gs1.best_estimator_
print(knn_best_params)
```

```python
# train scores

get_scores(gs1, X_train, y_train)
get_cm(gs1, X_train, y_train)
```

```python
# test scores and confusion matrix

get_scores(gs1, X_test, y_test)
get_cm(gs1, X_test, y_test)
```

**Evaluation/Comments:**

K-Nearest Neighbours is terribly overfitted, even after adjusting hyperparameters (increasing n_neighbors) to reduce overfitting. Hence, we will not be using this model.

#### Support Vector Machines

```python
# instatiate pipeline

svc_pipe = Pipeline([('preprocessor', preprocessor),
                     ('svc', LinearSVC(max_iter=100000))])
```

```python
# setting parameters

svc_params = {'svc__C': np.linspace(0.0001, 2, 10),
              'svc__penalty': ['l1', 'l2'],
              'svc__class_weight': ['balanced'],
              'svc__max_iter': [1000, 20000]}
```

```python
# instantiate gridsearch

gs2 = GridSearchCV(svc_pipe, param_grid = svc_params, cv = 5)
```

```python
# train model

gs2.fit(X_train, y_train)
```

```python
# get best hyperparams

svc_best_params = gs2.best_params_
svc_best_model = gs2.best_estimator_
print(svc_best_params)
```

```python
# train scores and confusion matrix

get_scores(gs2, X_train, y_train)
get_cm(gs2, X_train, y_train)
```

```python
# test scores and cm

get_scores(gs2, X_test, y_test)
get_cm(gs2, X_test, y_test)
```

**Evaluation/Comments:**

The train scores and test scores are very close, and does not show any under- or overfitting. The scores are also within the acceptable range (close to 90%). However, there are warnings for failure of convergence for certain hyperparameter values, which means that LinearSVM may not be the best model to use for this dataset.

### Feature Importance

Part of the impetus for this project is to help businesses identify which part of their business practices contribute the most to their sustainability/non-sustainability. As such, we will also look at which features contribute the most to the sustainability score.

```python
# get feature importance

best_classifier = rf_best_model.named_steps['rf']
feature_ipt = best_classifier.feature_importances_
feature_ipt
```

```python
# convert to df

feature_importance_dict = dict(zip(X.columns, feature_ipt))
sorted_feature_importance = sorted(feature_importance_dict.items(), key=lambda x: x[1], reverse=True)
importance = pd.DataFrame(sorted_feature_importance, columns=['feature', 'importance'])
importance
```

```python
# plot graph

import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(importance[:10], x = 'importance', y = 'feature', orient = 'h').set(title = 'Top 10 Features');
plt.xlabel('Feature importance')
plt.ylabel('Features');
```

**Comments:**

The features that contribute most to the sustainability score seem to be the material it is made from (regenerated, lyocell, viscose, etc.) and the volume of materials taken to make the garment (based on type: a t-shirt would require less material to manufacture than a dress, for example).

However, it is important to note that the features importance function in random forest only shows the impact of a feature on the target variable, and not whether the impact is negative or positive. As such, we are unable to infer if the use of more organic cotton, polyester, etc. in production will lead to a higher or lower sustainability score.

For example, while natural materials like organic cotton and cotton can be considered more sustainable because they are biodegradable, a lot of water is used to grow and produce cotton for weaving. Synthetic materials like polyester use less water to produce and can be easily made from recyled materials (other recycled polyester), but they are generally not biodegradable.
